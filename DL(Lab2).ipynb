```python
#step1 Given data
X=[0.75, 0.65, 0.80]  #feature
w=[0.4, 0.3, 0.5]
b=-0.2
Y=1
n=0.01
```

```python
print("Initial Weight: ",w)
print("Biased Weight: ",b)

```

```python
#Compute Net Input(z)=wixi+b
z=0
for i in range(len(X)):
  z+=w[i]*X[i]+b
  print(z)
z+=b
print("Net input: ",z)
```

```python
#Apply Activation Function
if(z>=0):
  y_pred=1
else:
  y_pred=0
print("predicted Output: ",y_pred)
```

```python
#Step4
Error=Y-y_pred
print("Error: ",Error)
```

```python
#STep5
if Error == 0:
  print("Prediction matches target, No Update required")
else:
  print("Wrong Prediction")

  # Step 6: Update weights
  for i in range(len(w)):
        w[i] = w[i] + n * Error * X[i]

  # Step 7: Update bias
  b = b + n * Error

# Step 8: Print updated results
print("\nUpdated Weights:", w)
print("Updated Bias:", b)
```
